```python
import pandas as pd
import numpy as np
```

```python
preprocessed_data=pd.read_csv("C:\\Users\\HP\\Downloads\\Absenteeism_preprocessed.csv")
```

```python
preprocessed_data.head()
```

```python
targets=np.where(preprocessed_data['Absenteeism Time in Hours']>(preprocessed_data['Absenteeism Time in Hours']).median(),1,0)
targets
```

```python
preprocessed_data['Excessive Absenteeism']=targets
preprocessed_data.head()
```

##A comment on the targets

```python
targets.sum()/targets.shape[0]
```

```python
data_with_targets=preprocessed_data.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets is preprocessed_data
```

```python
data_with_targets.shape
```

```python
unscaled_inputs=data_with_targets.iloc[:,:14]
unscaled_inputs.head()
```

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scalar= StandardScaler()
```

```python
absenteeism_scalar.fit(unscaled_inputs)
```

```python
scaled_inputs=absenteeism_scalar.transform(unscaled_inputs)
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,train_size=0.8,random_state=20)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

##Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```
